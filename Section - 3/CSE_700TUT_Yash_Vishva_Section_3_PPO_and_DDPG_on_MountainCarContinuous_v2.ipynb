```
class PPO:
    def __init__(self, ep, batch):

        self.ep = ep
        self.batch = batch

        self.env = gym.make('MountainCarContinuous-v0')
        self.bound = self.env.action_space.high[0]

        self.gamma = 0.9
        self.A_LR = 0.0001
        self.C_LR = 0.0002
        self.A_UPDATE_STEPS = 10
        self.C_UPDATE_STEPS = 10

        # Îµ for ppo
        self.epsilon = 0.2

        self.sess = tf.Session()
        self.build_model()

    def _build_critic(self):
        """critic model.
        """
        with tf.variable_scope('critic'):
            x = tf.layers.dense(self.states, 100, tf.nn.relu)

            self.v = tf.layers.dense(x, 1)
            self.advantage = self.dr - self.v

    def _build_actor(self, name, trainable):
        """actor model.
        """
        with tf.variable_scope(name):
            x = tf.layers.dense(self.states, 100, tf.nn.relu, trainable=trainable)

            mu = self.bound * tf.layers.dense(x, 1, tf.nn.tanh, trainable=trainable)
            sigma = tf.layers.dense(x, 1, tf.nn.softplus, trainable=trainable)

            norm_dist = tf.distributions.Normal(loc=mu, scale=sigma)

        params = tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope=name)

        return norm_dist, params

    def build_model(self):
        """build model with ppo loss.
        """
        # inputs
        self.states = tf.placeholder(tf.float32, [None, 2], 'states')
        self.action = tf.placeholder(tf.float32, [None, 1], 'action')
        self.adv = tf.placeholder(tf.float32, [None, 1], 'advantage')
        self.dr = tf.placeholder(tf.float32, [None, 1], 'discounted_r')

        # build model
        self._build_critic()
        nd, pi_params = self._build_actor('actor', trainable=True)
        old_nd, oldpi_params = self._build_actor('old_actor', trainable=False)

        # define ppo loss
        with tf.variable_scope('loss'):
            # critic loss
            self.closs = tf.reduce_mean(tf.square(self.advantage))

            # actor loss
            with tf.variable_scope('surrogate'):
                ratio = tf.exp(nd.log_prob(self.action) - old_nd.log_prob(self.action))
                surr = ratio * self.adv

                self.aloss = -tf.reduce_mean(tf.minimum(
                    surr,
                    tf.clip_by_value(ratio, 1.- self.epsilon, 1.+ self.epsilon) * self.adv))

        # define Optimizer
        with tf.variable_scope('optimize'):
            self.ctrain_op = tf.train.AdamOptimizer(self.C_LR).minimize(self.closs)
            self.atrain_op = tf.train.AdamOptimizer(self.A_LR).minimize(self.aloss)

        with tf.variable_scope('sample_action'):
            self.sample_op = tf.squeeze(nd.sample(1), axis=0)

        # update old actor
        with tf.variable_scope('update_old_actor'):
            self.update_old_actor = [oldp.assign(p) for p, oldp in zip(pi_params, oldpi_params)]


        self.sess.run(tf.global_variables_initializer())

    def choose_action(self, state):
        #choice continuous action from normal distributions.

        state = state[np.newaxis, :]
        action = self.sess.run(self.sample_op, {self.states: state})[0]

        return np.clip(action, -self.bound, self.bound)

    def get_value(self, state):
        #gets q value.

        if state.ndim < 2: state = state[np.newaxis, :]

        return self.sess.run(self.v, {self.states: state})

    def discount_reward(self, states, rewards, next_observation):
        #Computes target value.

        s = np.vstack([states, next_observation.reshape(-1, 2)])
        q_values = self.get_value(s).flatten()

        targets = rewards + self.gamma * q_values[1:]
        targets = targets.reshape(-1, 1)

        return targets

    def update(self, states, action, dr):
        #update the model

        self.sess.run(self.update_old_actor)

        adv = self.sess.run(self.advantage,
                            {self.states: states,
                             self.dr: dr})

        # update actor
        # run ppo2 loss
        for _ in range(self.A_UPDATE_STEPS):
            self.sess.run(self.atrain_op,
                          {self.states: states,
                            self.action: action,
                            self.adv: adv})

        # update critic
        for _ in range(self.C_UPDATE_STEPS):
            self.sess.run(self.ctrain_op,
                          {self.states: states,
                           self.dr: dr})

    def train(self):
        """train method.
        """
        tf.reset_default_graph()

        scores =[]
        window = 5
        for i in range(self.ep):
            observation = self.env.reset()

            states, actions, rewards = [], [], []
            episode_reward = 0
            j = 0

            while True:
                a = self.choose_action(observation)
                next_observation, reward, done, _ = self.env.step(a)
                states.append(observation)
                actions.append(a)

                episode_reward += reward
                rewards.append((reward + 8) / 8)

                observation = next_observation

                if (j + 1) % self.batch == 0:
                    states = np.array(states)
                    actions = np.array(actions)
                    rewards = np.array(rewards)
                    d_reward = self.discount_reward(states, rewards, next_observation)

                    self.update(states, actions, d_reward)

                    states, actions, rewards = [], [], []

                if done:
                    break
                j += 1

            scores.append(episode_reward)    
            ppo_score_history.append(episode_reward)
            if i % 100 == 0:
                df = pd.DataFrame(scores, columns=['r'])
                rolling_mean = df['r'].rolling(window=window).mean()
                plt.plot(rolling_mean)
                window = window + 15
                plt.show() 
 
            print('Episode: {} | Episode reward: {:.2f}'.format(i, episode_reward))


if __name__ == '__main__':
    ppo_score_history = []
    model = PPO(1501, 32)
    model.train()
```

```
import gym
import numpy as np
import random
import matplotlib.pyplot as plt

#neural-networks
import tensorflow.compat.v1 as tf
tf.compat.v1.disable_eager_execution()
import keras
from keras.models import Model, Sequential
from keras.layers import Dense, Input
from keras import backend as K
from keras.optimizers import Adam
from collections import deque
```

```
class ReplayBuffer(object):
    def __init__(self, buffer_size, input_shape, n_actions):
        self.buffer_size = buffer_size
        self.count=0
        #random.seed(10)
        self.states = np.zeros((self.buffer_size, *input_shape))
        self.next_state = np.zeros((self.buffer_size, *input_shape))
        self.actions = np.zeros((self.buffer_size, n_actions))
        self.rewards = np.zeros(self.buffer_size)
        self.terminals = np.zeros(self.buffer_size, dtype=np.float32)


    def store_transition(self, s, a, r, s2, t):
        index=self.count%self.buffer_size
        self.states[index]=s
        self.actions[index]=a
        self.rewards[index]=r
        self.next_state[index]=s2
        self.terminals[index]=1-t
        self.count += 1

    def sample_buffer(self, batch_size):
        if self.count < self.buffer_size:
            final_size = self.count
        else:
            final_size = self.buffer_size
        batch = np.random.choice(final_size, batch_size)


        states_ = self.states[batch]
        actions_ = self.actions[batch]
        rewards_ = self.rewards[batch]
        next_state_ = self.next_state[batch]
        terminals_ = self.terminals[batch]

        return states_, actions_, rewards_, next_state_, terminals_
```

```
class OrnsteinUhlenbeckActionNoise:
    def __init__(self, mu, sigma=0.15, theta=.2, dt=1e-2, x0=None):
        self.theta = theta
        self.mu = mu
        self.sigma = sigma
        self.dt = dt
        self.x0 = x0
        self.reset()

    def __call__(self):
        x = self.x_prev + self.theta * (self.mu - self.x_prev) * self.dt + \
                self.sigma * np.sqrt(self.dt) * np.random.normal(size=self.mu.shape)
        self.x_prev = x
        return x

    def reset(self):
        self.x_prev = self.x0 if self.x0 is not None else np.zeros_like(self.mu)

    def __repr__(self):
        return 'OrnsteinUhlenbeckActionNoise(mu={}, sigma={})'.format(self.mu, self.sigma)
```

```

class Actor(object):
    def __init__(self, learning_rate, action_size, name, input_size, sess, action_high):
        self.learning_rate = learning_rate
        self.action_size = action_size
        self.name = name
        self.input_size = input_size
        self.batch_size = 64
        self.sess = sess
        self.action_high = action_high

        ############# Actor Network ###############
        self.actor_brain()
        self.actor_network_params = tf.trainable_variables(scope=self.name)
        self.saver = tf.train.Saver()

        # Combine the gradients here
        self.actor_unnormalized_actor_gradients = tf.gradients(self.scaled_ouput, self.actor_network_params, -self.action_gradient)
        self.actor_actor_gradients = list(map(lambda x: tf.div(x, self.batch_size),self.actor_unnormalized_actor_gradients))

        # Optimization Op
        self.actor_optimize = tf.train.AdamOptimizer(self.learning_rate).apply_gradients(zip(self.actor_actor_gradients, self.actor_network_params))            

    def actor_brain(self):
        with tf.variable_scope(self.name):
            self.input_layer = tf.placeholder(tf.float32,shape=[None, *self.input_size],name='inputs')
            self.action_gradient = tf.placeholder(tf.float32,shape=[None, self.action_size],name='gradients')
            f1 = 1. / np.sqrt(800)
            dense1 = tf.layers.dense(self.input_layer, units=800,kernel_initializer=tf.initializers.random_uniform(-f1, f1),bias_initializer=tf.initializers.random_uniform(-f1, f1))
            batch1 = tf.layers.batch_normalization(dense1)
            layer1_activation = tf.nn.relu(batch1)
            f2 = 1. / np.sqrt(600)
            dense2 = tf.layers.dense(layer1_activation, units=600,kernel_initializer=tf.initializers.random_uniform(-f2, f2),bias_initializer=tf.initializers.random_uniform(-f2, f2))
            batch2 = tf.layers.batch_normalization(dense2)
            layer2_activation = tf.nn.relu(batch2)
            f3 = 0.003
            ouput = tf.layers.dense(layer2_activation, units=self.action_size,activation='tanh',kernel_initializer= tf.initializers.random_uniform(-f3, f3),bias_initializer=tf.initializers.random_uniform(-f3, f3))
            self.scaled_ouput = tf.multiply(ouput, self.action_high)
        
    def predict(self, inputs):
        return self.sess.run(self.scaled_ouput, feed_dict={self.input_layer: inputs})

    def train(self, inputs, gradients):
        self.sess.run(self.actor_optimize,
                      feed_dict={self.input_layer: inputs,
                                 self.action_gradient: gradients})
```

```
class Critic(object):
    def __init__(self, learning_rate, action_size, name, input_size, sess, action_high):
        self.learning_rate = learning_rate
        self.action_size = action_size
        self.name = name
        self.input_size = input_size
        self.batch_size = 64
        self.sess = sess
        self.action_high = action_high

        ############# Critic Network ###############
        self.critic_brain()
        self.critic_network_params = tf.trainable_variables(scope=self.name)

        # Optimization Op
        self.critic_optimize = tf.train.AdamOptimizer(self.learning_rate).minimize(self.critic_loss)

        # Combine the gradients here
        self.critic_actor_gradients = tf.gradients(self.critic_out, self.action_)

    def critic_brain(self):
        with tf.variable_scope(self.name):
            self.input_layer = tf.placeholder(tf.float32,shape=[None, *self.input_size],name='inputs')
            self.action_ = tf.placeholder(tf.float32,shape=[None, self.action_size],name='actions')
            self.q_target = tf.placeholder(tf.float32,shape=[None,1],name='targets')

            f1 = 1. / np.sqrt(800)
            dense1 = tf.layers.dense(self.input_layer, units=800,kernel_initializer=tf.initializers.random_uniform(-f1, f1),bias_initializer=tf.initializers.random_uniform(-f1, f1))
            batch1 = tf.layers.batch_normalization(dense1)
            layer1_activation = tf.nn.relu(batch1)
            f2 = 1. / np.sqrt(600)
            dense2 = tf.layers.dense(layer1_activation, units=600,kernel_initializer=tf.initializers.random_uniform(-f2, f2),bias_initializer=tf.initializers.random_uniform(-f2, f2))
            batch2 = tf.layers.batch_normalization(dense2)
            action_in = tf.layers.dense(self.action_, units=600, activation='relu')
            state_actions = tf.add(batch2, action_in)
            state_actions = tf.nn.relu(state_actions)
            f3 = 0.003
            self.critic_out = tf.layers.dense(state_actions, units=1,kernel_initializer= tf.initializers.random_uniform(-f3, f3),bias_initializer=tf.initializers.random_uniform(-f3, f3),kernel_regularizer=tf.keras.regularizers.l2(0.01))
            self.critic_loss = tf.losses.mean_squared_error(self.q_target, self.critic_out)
  
    def predict(self, inputs, actions):
        return self.sess.run(self.critic_out,
                             feed_dict={self.input_layer: inputs,
                                        self.action_: actions})
    def train(self, inputs, actions, q_target):
        return self.sess.run(self.critic_optimize,
                      feed_dict={self.input_layer: inputs,
                                 self.action_: actions,
                                 self.q_target: q_target})

    def get_action_gradients(self, inputs, actions):
        return self.sess.run(self.critic_actor_gradients,
                             feed_dict={self.input_layer: inputs,
                                        self.action_: actions})
```

```
class DDPGAgent(object):

    def __init__(self,input_size,action_size,action_high):
        
        self.sess = tf.Session()
        self.input_size = input_size
        self.action_size = action_size
        self.action_high = action_high
        self.learning_rate = 0.00005
        self.beta_learning_rate = 0.0005
        self.gamma = 0.99
        self.tau = 0.001
        self.mini_batch_size = 64
        self.memory = ReplayBuffer(10000000, input_size, action_size)

        #initilize Actor and critic
        self.actor = Actor(self.learning_rate, self.action_size, 'Actor', self.input_size, self.sess, self.action_high)
        self.critic = Critic(self.beta_learning_rate, self.action_size, 'Critic', self.input_size, self.sess, self.action_high)

        #initilize Target Actor and critic
        self.target_actor = Actor(self.learning_rate, self.action_size, 'TargetActor', self.input_size, self.sess, self.action_high)
        self.target_critic = Critic(self.beta_learning_rate, self.action_size, 'TargetCritic', self.input_size, self.sess, self.action_high)

        #get noise based on OU Noise function
        self.noise = OrnsteinUhlenbeckActionNoise(mu=np.zeros(action_size))

        #update critic and actor weights
        self.update_critic =  [self.target_critic.critic_network_params[i].assign( tf.multiply(self.critic.critic_network_params[i], self.tau) \
                    + tf.multiply(self.target_critic.critic_network_params[i], 1. - self.tau)) for i in range(len(self.target_critic.critic_network_params))]

        self.update_actor = [self.target_actor.actor_network_params[i].assign( tf.multiply(self.actor.actor_network_params[i], self.tau) \
                    + tf.multiply(self.target_actor.actor_network_params[i], 1. - self.tau)) for i in range(len(self.target_actor.actor_network_params))]

        self.sess.run(tf.global_variables_initializer())

        #update the neural net parameters
        self.old_tau = self.tau
        self.tau = 1.0
        self.target_critic.sess.run(self.update_critic)
        self.target_actor.sess.run(self.update_actor)        
        self.tau = self.old_tau
  
    #fine the probablities of action with noise included  
    def perfrom_action(self, current_state):
        state = current_state[np.newaxis, :]
        probabilities = self.actor.predict(state)
        noise = self.noise()
        action = probabilities + noise

        return action[0]

    def train(self):
        #replay buffer
        state, action, reward, new_state, done =  self.memory.sample_buffer(self.mini_batch_size)

        #train critic
        critic_prediction = self.target_critic.predict(new_state, self.target_actor.predict(new_state))
        target = []
        for j in range(self.mini_batch_size):
            target.append(reward[j] + self.gamma*critic_prediction[j]*done[j])
        target = np.reshape(target, (self.mini_batch_size, 1))
        self.critic.train(state, action, target)

        #train actor
        actor_prediction = self.actor.predict(state)
        grads = self.critic.get_action_gradients(state, actor_prediction)
        self.actor.train(state, grads[0])

        #update the neural net parameters
        self.target_critic.sess.run(self.update_critic)
        self.target_actor.sess.run(self.update_actor)

    #store experience for replay
    def remember(self, state, action, reward, new_state, done):
        self.memory.store_transition(state, action, reward, new_state, done)
```

```
#initialize gym environment and the agent
env = gym.make('LunarLanderContinuous-v2')
action_size = env.action_space.shape[0]
state_size = env.observation_space.shape[0]
#max action value
action_high=env.action_space.high
print(action_size,state_size,action_high)
agent = DDPGAgent([state_size],action_size,action_high)
score_history = []
loss_=[]
for i in range(360):
    obs = env.reset()
    done = False
    score = 0
    
    while not done:
        #action
        act = agent.perfrom_action(obs)
        new_state, reward, done, info = env.step(act)
        #experience replay
        agent.remember(obs, act, reward, new_state, int(done))
        #train
        agent.train()
        score += reward
        obs = new_state
    score_history.append(score)
    print("Episode ",i)
    print('score:' % score,'avg: ',np.mean(score_history[-100:]))
     
plt.plot(score_history)
plt.title('Rewards per epoch')
plt.show()
```
